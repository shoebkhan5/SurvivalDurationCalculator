```python
class LifetimeCalculator:
    def __init__(self, age_in_years):
        self.age_in_years = age_in_years

    def get_age_in_months(self):
        "Calculate age in months based on the given age in years."
        return self.age_in_years * 12

    def get_age_in_weeks(self):
        "Calculate age in weeks based on the given age in years."
        return self.age_in_years * 52

    def get_age_in_days(self):
        "Calculate age in days based on the given age in years, assuming 365 days per year."
        return self.age_in_years * 365

    def get_age_in_hours(self):
        "Calculate age in hours based on the given age in years."
        return self.age_in_years * 365 * 24

    def get_age_in_minutes(self):
        "Calculate age in minutes based on the given age in years."
        return self.age_in_years * 365 * 24 * 60

    def get_age_in_seconds(self):
        "Calculate age in seconds based on the given age in years."
        return self.age_in_years * 365 * 24 * 60 * 60

def request_time_unit():
    "Get valid time unit input from user."
    unit = input("Choose a time unit (Months, Weeks, Days, Hours, Minutes, Seconds): ").lower()
    valid_units = ['m', 'months', 'w', 'weeks', 'd', 'days', 'h', 'hours', 'min', 'minutes', 's', 'seconds']
    
    if unit not in valid_units:
        print("Invalid input! Please choose a valid time unit.")
        return request_time_unit()  
    return unit

def run_program():
    "Main program to calculate survival duration based on age and chosen unit."
    try:
        age_in_years = int(input("Enter your age: "))
        if age_in_years < 0:
            print("Age cannot be negative! Please enter a valid age.")
            return
        
        
        unit = request_time_unit()

        
        calculator = LifetimeCalculator(age_in_years)

        
        if unit in ['m', 'months']:
            result = calculator.get_age_in_months()
        elif unit in ['w', 'weeks']:
            result = calculator.get_age_in_weeks()
        elif unit in ['d', 'days']:
            result = calculator.get_age_in_days()
        elif unit in ['h', 'hours']:
            result = calculator.get_age_in_hours()
        elif unit in ['min', 'minutes']:
            result = calculator.get_age_in_minutes()
        elif unit in ['s', 'seconds']:
            result = calculator.get_age_in_seconds()

        
        print(f"You have lived for {result} {unit.capitalize()}.")

    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    run_program()
    
```
